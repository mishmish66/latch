```python
from matplotlib import pyplot as plt

from latch.models.nets import FreqLayer

from jax import numpy as jnp
```

```python
n = 4
d = 32

fl = FreqLayer(d, min_freq=0.001)

coded = fl(jnp.linspace(0, 100, n))

for i in range(n):
    plt.plot(coded[i * d // n : (i + 1) * d // n].real, label=f"Real {i}")
```
