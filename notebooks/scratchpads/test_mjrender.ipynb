```python
import os
os.environ["MUJOCO_GL"] = "EGL"

import mujoco
```

```python
model = mujoco.MjModel.from_xml_path("../../latch/env/finger/assets/finger.xml")
```

```python
model
```

```python
renderer = mujoco.Renderer(model)
```

```python
renderer.update_scene(data, "topdown")
```
